```python
import pandas as pd
import numpy as np
```

```python
dfmain = pd.read_csv('soccer17.csv', index_col = False, usecols =['Div','Y','HomeTeam','AwayTeam','FTHG','FTAG'])
```

# Question 1A

```python
epl_17 = dfmain.loc[(dfmain['Div'] == 'EPL') & (dfmain['Y'] == 17)]
epl_17['GD'] = epl_17['FTHG'] - epl_17['FTAG']
```

```python
df = pd.DataFrame({'GamesPlayed': (epl_17.groupby('HomeTeam').size() + epl_17.groupby('AwayTeam').size())}).reset_index()
df = df.rename(columns={'HomeTeam': 'Team'})
```

```python
gd = np.empty(len(df))
w = np.empty(len(df))
d = np.empty(len(df))
for i in range(len(df)):
    team = df.Team.iat[i]
    gs = epl_17.loc[epl_17.HomeTeam == team, 'FTHG'].sum() + epl_17.loc[epl_17.AwayTeam == team, 'FTAG'].sum()
    gc = epl_17.loc[epl_17.HomeTeam == team, 'FTAG'].sum() + epl_17.loc[epl_17.AwayTeam == team, 'FTHG'].sum()
    wins = len(epl_17.loc[(epl_17.HomeTeam == team) & (epl_17.GD > 0)]) \
    + len(epl_17.loc[(epl_17.AwayTeam == team) & (epl_17.GD < 0)])
    draws = len(epl_17.loc[(epl_17.HomeTeam == team) & (epl_17.GD == 0)]) \
    + len(epl_17.loc[(epl_17.AwayTeam == team) & (epl_17.GD == 0)])
    gd[i] = gs-gc
    w[i] = wins
    d[i] = draws
df['AverageGD'] = gd/df['GamesPlayed']
df['Wins'] = w
df['Draws'] = d
df['Losses'] = df['GamesPlayed'] - df['Wins'] - df['Draws']
df = df.sort_values(by=['AverageGD'], ascending = False)
df
```

# Question 1B

```python
df['Points'] = 3*df['Wins'] + 1*df['Draws']
df = df.sort_values(by=['Points'], ascending = False)
df
```
